```python
import rospy

from sensor_msgs.msg import LaserScan
from geometry_msgs.msg import Twist

class Roamer:

    def __init__(self):
        self.laser_sub = rospy.Subscriber("/scan",
                                          LaserScan,
                                          self.callback)
        self.twist_pub = rospy.Publisher('/mobile_base/commands/velocity', Twist, queue_size=1)

    def callback(self, data):
        min_range = data.range_max
        print len(data.ranges)
        for v in data.ranges:
            if v < min_range:
                min_range = v
        print min_range
        twist = Twist()
        if min_range < 1:
            twist.angular.z = 1
        else:
            twist.linear.x = 0.2
        self.twist_pub.publish(twist)
            
        

ic = Roamer()
rospy.init_node('Roamer')
rospy.spin()
```
